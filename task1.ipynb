**Goal:** Load, inspect, and analyze the Iris dataset distributions and outliers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
print("Loading Iris Dataset...")
df = sns.load_dataset('iris')

# Inspect structure
print(f"Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")
print("\n--- First 5 Rows ---")
print(df.head())
```

```python
print("--- Data Structure & Types ---")
df.info()

print("\n--- Summary Statistics ---")
print(df.describe())
```

```python
# A. Scatter Plot Matrix
sns.pairplot(df, hue='species', palette='Set1')
plt.suptitle("Feature Relationships by Species", y=1.02)
plt.show()

# B. Histograms
df.hist(figsize=(10, 6), bins=15, edgecolor='black', color='teal')
plt.suptitle("Feature Value Distributions")
plt.tight_layout()
plt.show()

# C. Box Plots for Outlier Detection
plt.figure(figsize=(10, 5))
sns.boxplot(data=df.drop(columns=['species']), palette="Pastel1")
plt.title("Box Plots: Outlier Identification")
plt.ylabel("Measurement (cm)")
plt.show()
```

### Key Insights & Observations:
1. **Species Separation:** In the scatter plots, the `setosa` species is distinctly separated from `versicolor` and `virginica`.
2. **Outliers:** The Box Plot shows that `sepal_width` has a few data points lying outside the whiskers, indicating minor outliers.
3. **Distributions:** The histograms show that petal dimensions (`petal_length` and `petal_width`) have a clear gap in their distributions, which aligns with how easily the `setosa` species is classified.
